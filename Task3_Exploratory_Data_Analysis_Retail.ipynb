## EXPLORATORY ANALYSIS

```python
# Draw count of each states
plt.figure(figsize=(15,8))
sns.countplot(x='State',data=superstore_df,order=superstore_df['State'].value_counts().index)
plt.xticks(rotation=60)
plt.show()
```

### Insights - 
#### Helps in getting that which state have the highest selling products of our company and where we need to do more marketing ,so that everybody in that region is going to know about our company.


```python
# Draw count of Category

plt.figure(figsize=(8,5))
sns.countplot(x='Category',data=superstore_df,palette='spring_r',order=superstore_df['Category'].value_counts().index)
plt.xticks(rotation=30)
plt.show()
```

```python
# Draw Count of Sub-category 

plt.figure(figsize=(15,12))
sns.countplot(x='Sub-Category',data=superstore_df,palette='spring_r',order=superstore_df['Sub-Category'].value_counts().index)
plt.xticks(rotation=60)
plt.show()
```

### Insights
#### 1)Our best selling product categories are Blinders and paper follwed by furnishing, phones and others.
#### 2)Copiers and Machines are the least sold products.

```python
# Display the dataset contains states, sales and profit columns

profit_df = superstore_df.groupby("State")[["Sales", "Profit"]].sum()
profit_df
```

```python
# Draw a Bar chart of sales vs profit of each states

profit_df.plot.bar(figsize=(15,8))
```

### Insights- 
#### 1) California makes the highest profit followed by New York and Washington
#### 2) We also see that Texas, Ohio and Pennsylvania are running in loss

```python
# Display Count of Sub-category, Discount and Profit columns.

sc_disc = superstore_df.groupby("Sub-Category")[["Discount","Profit"]].sum()
sc_disc
```

```python
# Draw Scatter plot of Dicount VS Profit

superstore_df.plot.scatter(x='Discount',y='Profit',color='black',marker='.',alpha=0.4)
```

### Insight-
#### 1)From the graph we get to kow that when there is less discount we have more profit but when we gives high discount then we got less profit.

```python
# Draw Bar Chart of Sales VS Profit of Sub-category

sc_pro = superstore_df.groupby("Sub-Category")[["Profit", "Sales"]].sum()
sc_pro.plot.bar(figsize=(20,10))
```
